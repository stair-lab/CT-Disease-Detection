```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

## Test Set

```python
df = pd.read_csv('/lfs/turing4/0/mahmedc/CTdata/test.csv', sep = ',')
```

```python
# Plot the frequency distribution: Log Transform
plt.figure(figsize=(8, 6))
plt.hist(np.log10(df['CTBiomarkers.CalciumScoring.AbdominalAgatston'] + 1e-6), 
         bins=range(int(min(np.log10(df['CTBiomarkers.CalciumScoring.AbdominalAgatston'] + 1e-6))), int(max(np.log10(df['CTBiomarkers.CalciumScoring.AbdominalAgatston'] + 1e-6))) + 2), 
         edgecolor='black', width = 2.0)


# Add labels and title
plt.title('Frequency Distribution of CTBiomarkers.CalciumScoring.AbdominalAgatston')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show plot
plt.show()
```

```python
# Plot the frequency distribution: Inverse Transform
plt.figure(figsize=(8, 6))
plt.hist(1/(df['CTBiomarkers.CalciumScoring.AbdominalAgatston'] + 1e-6), 
         bins=range(int(min(1/(df['CTBiomarkers.CalciumScoring.AbdominalAgatston'] + 1e-6))), int(max(1/(df['CTBiomarkers.CalciumScoring.AbdominalAgatston'] + 1e-6))) + 2), 
         edgecolor='black', width = 2.0)


# Add labels and title
plt.title('Frequency Distribution of CTBiomarkers.CalciumScoring.AbdominalAgatston')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show plot
plt.show()
```

```python
# Summary statistics for a specific column
summary = df['CTBiomarkers.CalciumScoring.AbdominalAgatston'].describe()

print(summary)
```

```python
# Min-Max normalization of 'column_name'
min_val = df['CTBiomarkers.CalciumScoring.AbdominalAgatston'].min()
max_val = df['CTBiomarkers.CalciumScoring.AbdominalAgatston'].max()

df['CalciumScoreNormalized'] = (df['CTBiomarkers.CalciumScoring.AbdominalAgatston'] - min_val) / (max_val - min_val)
```

```python
# Plot the frequency distribution
# Define the bin edges
bins = [0.0, 0.10, 0.20, 0.30, 0.40, 0.50, 0.60, 0.70, 0.80, 0.90, 1.00]

new_bins = [x/10 for x in bins]

# Plot the bucketed frequency distribution
plt.hist(df['CalciumScoreNormalized'][df['CalciumScoreNormalized'] <= 0.10], bins=new_bins, edgecolor='black')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Frequency Distribution of CalciumScoreNormalized')

# Show the plot
plt.show()
```

```python
# Plot the frequency distribution: Inverse Transform
plt.figure(figsize=(8, 6))
plt.hist(1/(df['CalciumScoreNormalized'] + 1e-6), 
         bins=range(int(min(1/(df['CalciumScoreNormalized'] + 1e-6))/10), int(max(1/(df['CalciumScoreNormalized'] + 1e-6)))/10 + 2), 
         edgecolor='black', width = 2.0)


# Add labels and title
plt.title('Frequency Distribution of Inverse Transform CalciumScoreNormalized')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show plot
plt.show()
```

```python
# Create a histogram
plt.figure(figsize=(8, 6))
plt.hist(df['CalciumScoreNormalized'], 
         bins=range(int(min(df['CalciumScoreNormalized'])), 
                    int(max(df['CalciumScoreNormalized'])) + 2), edgecolor='black')

# Add labels and title
plt.title('Frequency Distribution of CalciumScoreNormalized')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show plot
plt.show()
```

```python
# Summary statistics for a specific column
summary = df['CalciumScoreNormalized'].describe()

print(summary)
```

```python
df['CalciumScoreInverted'] = (1/(df['CTBiomarkers.CalciumScoring.AbdominalAgatston'] + 1e-6))
summary = df['CalciumScoreInverted'].describe()

print(summary)
```

## Image Visualization

```python
from torchvision import transforms
from PIL import Image
```

```python
img = Image.open('/lfs/turing4/0/mahmedc/CTdata/data/010119-0311.png')
img = img.resize((256,256), Image.LANCZOS)
img = img.convert('L')
img.show()
```

```python
img
```

```python
train_transform = transforms.Compose([
        transforms.RandomHorizontalFlip(p=0.2),
        transforms.RandomApply([
            transforms.RandomAffine(10),
            transforms.RandomResizedCrop(256, scale=(1.0, 1.1), ratio=(0.75, 1.33)),
            transforms.ColorJitter(brightness=0.2, contrast=0.2)
        ], p=0.75),
        transforms.RandomPerspective(distortion_scale=0.2, p=0.75),
        # transforms.ToTensor(),
        # transforms.Normalize((0.55001191,), (0.18854326,))
    ])
```

```python
img_transformed = train_transform(img)
```
